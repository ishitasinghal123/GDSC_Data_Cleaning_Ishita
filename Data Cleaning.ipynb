```python
bd = df["R&D Spend"].quantile(0.02)
td = df["R&D Spend"].quantile(0.98)

bdA = df["Administration"].quantile(0.02)
tdA = df["Administration"].quantile(0.98)

bdB = df["Marketing Spend"].quantile(0.02)
tdB = df["Marketing Spend"].quantile(0.98)

bdC = df["Marketing Spend"].quantile(0.02)
tdC = df["Profit"].quantile(0.98)

s1 = df[(df["R&D Spend"] > bd)
                & (df["R&D Spend"] < td)]
s2 = s1[(s1["Administration"] > bdA)
                & (s1["Administration"] < tdA)]

s3 = s2[(s2["Marketing Spend"] > bdB)
                & (s2["Marketing Spend"] < tdB)]

s4 = s3[(s3['Marketing Spend'] > bdC)
                & (s3["Marketing Spend"] < tdC)]
s4
```
